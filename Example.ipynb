```python
from datasets import make_regression, print_shape, sort_dataset
from plots import plot2D, plot3D
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=1000, n_features=1, noise=0.0, random_state=111)
```

```python
print_shape(X, y)
```

```python
X1, y1 = sort_dataset(X, y)
```

```python
title = "2D Syntactic Regression Dataset"
xlabel = "Feature 1"
ylabel = "Target 1"
plot2D(title, xlabel, ylabel, X1, y1)
```
